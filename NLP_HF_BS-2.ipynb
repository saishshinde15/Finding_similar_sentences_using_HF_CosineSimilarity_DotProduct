# Creating a array of models and a loop to ruun. each model for cosine simailarty and dot product to save compuationals efforts and time

```python
models=[
    'sentence-transformers/paraphrase-MiniLM-L6-v2',
    'sentence-transformers/all-mpnet-base-v2',
    'sentence-transformers/paraphrase-mpnet-base-v2',
    'sentence-transformers/all-distilroberta-v1',
    'sentence-transformers/all-roberta-large-v1',
    'sentence-transformers/paraphrase-distilroberta-base-v1',
    'sentence-transformers/nli-mpnet-base-v2',
    'sentence-transformers/nli-roberta-base-v2',
    'sentence-transformers/stsb-roberta-large',
    'sentence-transformers/all-MiniLM-L12-v2'
]
```

```python
for modelname in models:
  model=SentenceTransformer(modelname)
  #print(modelname)
  print(f"\nModel: {modelname}")

  embedding1=model.encode(sentence1,convert_to_tensor=True)
  embedding2=model.encode(sentence2,convert_to_tensor=True)

  cosine_similarity_1=util.pytorch_cos_sim(embedding1[0],embedding1[1]).item()
  cosine_similarity_2=util.pytorch_cos_sim(embedding2[0],embedding2[1]).item()

  dot_product1=util.dot_score(embedding1[0],embedding1[1]).item()
  dot_product2=util.dot_score(embedding2[0],embedding2[1]).item()

  print(f"Cosine Similarity 1: {cosine_similarity_1}")
  print(f"Cosine Similarity 2: {cosine_similarity_2}")

  print(f"Dot Product 1: {dot_product1}")
  print(f"Dot Product 2: {dot_product2}")
```
